# Tratamento de Dados


  ## Análise Exploratória, verificar características
  

```python
df
```

```python
europe = df[df.continent=='Europe']
```

```python
europe.reset_index(drop=True, inplace=True)
europe
```

```python
df[df.iso_code=='GEO']
```

Selecionando colunas para o tratamento

```python
europe = europe[['iso_code', 'location', 'date', 'total_cases', 'new_cases', 'total_deaths', 'new_deaths','new_tests', 'total_tests', 'total_vaccinations', 'people_vaccinated',
'people_fully_vaccinated', 'population', 'population_density', 'median_age', 'life_expectancy', 'human_development_index' ]].copy()
```

## Limpeza e Preparação dos dados


**Tratando os dados ausentes**

```python
europe.isnull().sum()
```

```python
europe.fillna({'total_cases': 0, 
               'new_cases': 0,
               'new_deaths': 0, 
               'total_deaths': 0,
               'new_tests': 0,
               'total_tests': 0, 
               'total_vaccinations': 0,
               'people_vaccinated': 0, 
               'people_fully_vaccinated': 0,
               'population_density': -1,
               'median_age': -1, 
               'life_expectancy': -1,
               'human_development_index':-1
                }, inplace = True) 
```

**Convertendo o tipo das datas em String para Datetime**

```python
europe.date = pd.to_datetime(europe.date)
```

**Verificando a existência de dados duplicados**

```python
print(europe.duplicated())
print("\nQuantidade de duplicatas:", europe.duplicated().sum())
```

## Junção, Combinação e reformatação



Paises liderados por mulheres

```python
europe_fem = europe.loc[europe.location.isin(['Germany', 'Belgium', 'Croatia', 'Denmark', 'Estonia', 'Slovakia', 'Finland', 'Norway', 'Iceland', 'Serbia', 'Greece'])]
europe_fem.reset_index(drop=True, inplace=True) 
europe_fem
```

```python
europe_fem.describe()
```

```python
europe_fem = europe_fem[europe_fem.new_deaths >= 0]
```

```python
europe_fem = europe_fem[europe_fem.new_cases >= 0]
```

```python
europe_fem = europe_fem[europe_fem.new_tests >= 0]
```

Paises liderados por homens


```python
europe_masc = europe.loc[~europe.location.isin(['Germany', 'Belgium', 'Croatia', 'Denmark', 'Estonia', 'Slovakia', 'Finland', 'Norway', 'Iceland', 'Serbia', 'Greece'])] 
europe_masc.reset_index(drop=True, inplace=True)
europe_masc
```

# Visualização de dados com Storytelling

Em 31 de dezembro de 2019, a Organização Mundial da Saúde (OMS) foi alertada sobre vários casos de pneumonia na cidade de Wuhan, província de Hubei, na República Popular da China. Tratava-se de uma nova cepa (tipo) de coronavírus que não havia sido identificada antes em seres humanos.

Uma semana depois, em 7 de janeiro de 2020, as autoridades chinesas confirmaram que haviam identificado um novo tipo de coronavírus. Os coronavírus estão por toda parte. Eles são a segunda principal causa de resfriado comum (após rinovírus) e, até as últimas décadas, raramente causavam doenças mais graves em humanos do que o resfriado comum.

Em 30 de janeiro de 2020, a OMS declarou que o surto do novo coronavírus constitui uma Emergência de Saúde Pública de Importância Internacional (ESPII) – o mais alto nível de alerta da Organização, conforme previsto no Regulamento Sanitário Internacional. Essa decisão buscou aprimorar a coordenação, a cooperação e a solidariedade global para interromper a propagação do vírus. 

Cada país se tornou responsável pela segurança e proteção de suas respectivas populações, adotando medidas de contenção como o uso de mascaras, isolamento social, dentre outras. Analisaremos como essas medidas afetaram o desenvolvimento da covid19 nos países liderados por mulheres na Europa
  

## Visualização dos índices dos paises liderados por mulheres

Ao todo, 11 dos 51 países da Europa possuem uma liderança feminina. Antes de verificarmos o desenvolvimento da pandemia, visualizaremos alguns índices de desenvolvimento (IDH, Idade media, População, Expectativa de Vida) dos países da Europa.

**Índices dos países liderados por mulheres**

```python
indices_fem = europe_fem[['location','population', 'population_density', 'median_age', 'life_expectancy', 'human_development_index']].copy()
```

```python
indices_fem.drop_duplicates(['human_development_index'], keep='last')
```

**Indice de Desenvolvimento Humano**

```python
plt.figure(figsize=(12,6))
sns.barplot(x='location', y='human_development_index', data=indices_fem.sort_values(by='human_development_index', ascending=False), color='#006699')
plt.xlabel('')
plt.ylabel('IDH')
plt.title('Indice de Desenvolvimento Humano dos Países Liderados por Mulheres')
plt.show()
```

**Expectativa de Vida**

```python
plt.figure(figsize=(12,6))
sns.barplot(x='location', y='life_expectancy', data=indices_fem.sort_values(by='life_expectancy', ascending=False), color='#993399')
plt.xlabel('')
plt.ylabel('Idade')
plt.title('Expectativa de Vida dos Países Liderados por Mulheres')
plt.show()
```

**Idade Média**

```python
plt.figure(figsize=(12,6))
sns.barplot(x='location', y='median_age', data=indices_fem.sort_values(by='median_age', ascending=False), color='#006699')
plt.xlabel('')
plt.ylabel('Idade Média')
plt.title('Idade Média dos Países Liderados por Mulheres')
plt.show()
```

**População e Densidade da População**

```python
plt.figure(figsize=(12,6))
sns.barplot(x='location', y='population', data=indices_fem.sort_values(by='population', ascending=False), color='#993399')
plt.xlabel('')
plt.ylabel('População (Em milhões)')
plt.title('População dos Países Liderados por Mulheres')
plt.show()
```

```python
plt.figure(figsize=(12,6))
sns.barplot(x='location', y='population_density', data=indices_fem.sort_values(by='population', ascending=False), color='#cc66cc')
plt.xlabel('')
plt.ylabel('População / Km²')
plt.title('Densidade Demográfica dos Países Liderados por Mulheres')
plt.show()
```

## Avanço da Covid19 nos países liderados por mulheres na europa



**Total de Casos e Mortes em cada um dos países**

```python
fig, axes = plt.subplots(2,1, figsize=(12,24))

sns.set_theme(style="whitegrid")

sns.lineplot(ax=axes[0], data=europe_fem, x='date', y='total_cases', hue='location', palette="tab10", linewidth=2.5)
sns.lineplot(ax=axes[1], data=europe_fem, x='date', y='total_deaths', hue='location', palette="tab10", linewidth=2.5)

axes[0].set_xlabel('')
axes[0].set_ylabel('Total de Casos (Em milhões)')
axes[0].set_title('Total de Casos de Covid19 nos Países Liderados por Mulheres')


axes[1].set_xlabel('')
axes[1].set_ylabel('Total de Mortes')
axes[1].set_title('Total de Mortes por Covid19 nos Países Liderados por Mulheres')


plt.show()


```

**Somatório dos Casos e Mortes diárias desses países**

```python
fig, axes = plt.subplots(2,1, figsize=(12,18))

sns.set_theme(style="whitegrid")

sns.lineplot(ax=axes[0], data=europe_fem, x='date', y='new_cases', palette="tab10")
sns.lineplot(ax=axes[1], data=europe_fem, x='date', y='new_deaths', palette="tab10")

axes[0].set_xlabel('')
axes[0].set_ylabel('Total de Casos')
axes[0].set_title('Casos Diários de Covid19 nos Países Liderados por Mulheres')


axes[1].set_xlabel('')
axes[1].set_ylabel('Total de Mortes')
axes[1].set_title('Mortes Diárias por Covid19 nos Países Liderados por Mulheres')

plt.show()
```

**Densidade de Casos e Mortes Diárias**

```python
fig, axes = plt.subplots(2,1, figsize=(12,18))

sns.set_theme(style="whitegrid")

sns.boxplot(ax=axes[0], data=europe_fem, x='location', y='new_cases', palette="tab10")
sns.boxplot(ax=axes[1], data=europe_fem, x='location', y='new_deaths', palette="tab10")

axes[0].set_xlabel('')
axes[0].set_ylabel('Total de Casos')
axes[0].set_title('Densidade de Casos Diários de Covid19 nos Países Liderados por Mulheres')


axes[1].set_xlabel('')
axes[1].set_ylabel('Total de Mortes')
axes[1].set_title('Densidade de Mortes Diárias  por Covid19 nos Países Liderados por Mulheres')


plt.show()

```
