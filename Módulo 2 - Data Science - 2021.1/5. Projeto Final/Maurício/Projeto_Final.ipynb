```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
url = 'https://raw.githubusercontent.com/Ramo-Estudantil-IEEE-UFBA/ArtIEEEficiais/main/M%C3%B3dulo%202%20-%20Data%20Science%20-%202021.1/5.%20Projeto%20Final/Covid-19-World-Dataset.csv'
df = pd.read_csv(url)
df
```

# Tratamento de Dados


  ## Análise Exploratória, verificar características
  

Através do seguinte comando, pôde-se observar a lista de continentes do Dataframe:

```python
df['continent'].unique()
```

Observa-se que existem elementos do tipo NaN, isto se deve ao fato de que existem informações no Dataframe que não são um país em específico, mas dados gerais do continente Africano, como irei realizar uma pequena análise nos dados do Brasil, isto não será um problema.

Para filtrar os dados que pertencem somente ao Brasil:

```python
brasil = df[df.iso_code=='BRA']
brasil
```

## Limpeza e Preparação dos dados


Para resetar o índice, poderia ser realizado o seguinte comando:

```python
brasil = brasil.reset_index(drop=True)
brasil
```

Como os dados são do Brasil, é redundante ter as colunas `iso_code`, `continent` e `location`. Para isso, devemos removê-las.

```python
brasil = brasil.drop(['iso_code','continent','location'], axis = 1) 
brasil
```

Para saber a quantidade de elementos nulos:

```python
brasil.isnull().sum()
```

Como realizaremos uma pequena e introdutória análise, iremos utilizar somente as columas `total_deaths`, `new_deaths`,`total_cases` e `new_cases` deste Dataframe, para isso, preenchemos os valores nulos de `total_deaths`, `new_deaths` com **0**:

```python
brasil.total_deaths = brasil.total_deaths.fillna(0)
brasil.new_deaths = brasil.new_deaths.fillna(0)
brasil
```

## Junção, Combinação e Reformatação



Primeiramente formataremos a data para um tipo de data, isso será muito útil quando estivermos visualizando séries temporais:

```python
brasil["date"]=pd.to_datetime(brasil["date"])
```

Após isso, renomearemos as colunas para camelCase:

```python
brasil = brasil.rename(columns={'date': 'data', 'total_cases': 'totalCasos','new_cases':'novosCasos',
                                'total_deaths':'totalMortes','new_deaths':'novasMortes'});
brasil
```

E finalmente recortamos a parte do Dataframe de nosso interesse na análise de dados:

```python
brasilRecorte = brasil[['data','totalCasos','novosCasos','totalMortes','novasMortes']]
brasilRecorte
```

# Visualização de dados com Storytelling
  

## Visualização dos dados de classificação


Numa tentativa de visualizar inicialmente a quantidade de infecções da doença, foi utilizado um gráfico de pizza, para se visualizar melhor a proporção do Brasil que foi infectada pelo vírus.

```python
casos = brasilRecorte.totalCasos.iloc[-1] 
infeccoes = pd.Series([casos,brasil.population.iloc[-1]-casos], 
                      index=['Total de infecções', 'População Brasileira'], name='')

plot = infeccoes.plot(kind='pie', title="Infecções COVID", 
                      autopct='%.2f%%', explode=(0, 0.1),
                      shadow=True)
```

Para se ter uma noção da distribuição de mortes por dia no pais:

```python
novasMortes = brasilRecorte.novasMortes
plot = novasMortes.plot(kind='box', showfliers=False)

novasMortes.describe()
```

O gráfico acima, juntamente com os dados numéricos, mostram que desde o primeiro caso de COVID19 no país, até o mês de junho de 2021, morreram em média, mais de 1000 pessoas por dia, chegando a dias com 4249 mortes.

Além disso, metade dos dias tivemos entre 480 a 1289 mortes.

## Visualização dos dados discretos


Para visualizar o distribuição da quantidade de mortes ao longo dos dias e utilizando um filtro de média móvel, temos: 

```python
novasMortes.plot(color='gray')
novasMortes.rolling(7).mean().plot(color="black")
```

Assim como para a quantidade de novos casos diários:

```python
novosCasos = brasilRecorte.novosCasos

novosCasos.plot(color='gray')
novosCasos.rolling(7).mean().plot(color="black")
```

## Visualizando séries temporais



Podemos utilizar a coluna `data` como índice do Dataframe:

```python
brasilTemporal = brasilRecorte.set_index('data')
brasilTemporal
```

Isso permite acessar ano, mês e dia, individualmente. Utilizamos para criar novas colunas.

```python
brasilTemporal["ano"] = brasilTemporal.index.year
brasilTemporal["mes"] = brasilTemporal.index.month
brasilTemporal["dia"] = brasilTemporal.index.day

brasilTemporal
```

Utilizando agora o seaborn, podemos ver os mesmos gráficos acima, agora com séries temporais

```python
sns.set(rc={'figure.figsize':(11, 4)})
brasilTemporal['novosCasos'].plot(color='gray')
brasilTemporal['novosCasos'].rolling(7).mean().plot(color="black")
```

```python
sns.set(rc={'figure.figsize':(11, 4)})
brasilTemporal['novasMortes'].plot(color='gray')
brasilTemporal['novasMortes'].rolling(7).mean().plot(color="black")
```
